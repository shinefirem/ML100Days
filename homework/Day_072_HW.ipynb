# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def relu(x):
    x = x.copy()
    x[x<0] = 0
    return x
def drelu(x):
    x = x.copy()
    x[x<0] = 0
    x[x>0] = 1
    return x
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, relu(x), 'b', label='linspace(-10,10,10)')
plt.grid()
plt.show()
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, drelu(x), 'b', label='linspace(-10,10,10)')
plt.grid()
plt.show()
```
